```python
import pandas as pd
import matplotlib.pyplot as plt


consumi = pd.read_csv('Global energy consumption.csv', header=0, 
                      names=['Entity', 'Code', 'Year', 'Wind', 'Oil', 'Nuclear', 'Hydropower', 'Bimass',
                             'Other renewables', 'Biofuels', 'Solar', 'Coal', 'Gas'])
```

```python
consumi.drop(['Entity', 'Code'], axis=1, inplace=True, errors='ignore')
```

```python
years = []
consumption = []

for index, row in consumi.iterrows():
    years.append(row['Year'])
    consumption.append(row['Wind'] + row['Oil'] + row['Nuclear'] + row['Hydropower'] + row['Bimass'] + 
                        row['Other renewables'] + row['Biofuels'] + row['Solar'] + row['Coal'] + row['Gas'])
```

```python
plt.figure(figsize=(16,10))
plt.title("Global energy consumption per year", fontsize=20, pad=10)

plt.plot(years, consumption, linewidth=2)
plt.grid(True)
plt.margins(x=0)
plt.ylabel("TWh", fontsize=15, labelpad=10)
plt.xticks(fontsize=14) 
plt.yticks(fontsize=14) 
plt.savefig('Global energy consumption per year.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
