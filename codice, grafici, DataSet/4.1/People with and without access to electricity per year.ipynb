```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('number-of-people-with-and-without-electricity-access.csv')

result = df[df['Entity'] == 'World']
result = result.drop(['Entity', 'Code'], axis=1)
```

```python
# preparo i font
color_red = '#fd625e'
color_blue = '#01b8aa'
column0 = result['Number of people without access to electricity']
column1 = result['Number of people with access to electricity']

# plot
fig, axes = plt.subplots(figsize=(12,8), ncols=2)
fig.tight_layout()

axes[0].barh(result.Year, column0, align='center', color=color_red, zorder=10)
axes[0].set_title('without access', fontsize=18, pad=15, color=color_red)
axes[1].barh(result.Year, column1, align='center', color=color_blue, zorder=10)
axes[1].set_title('with access', fontsize=18, pad=15, color=color_blue)

axes[0].invert_xaxis()
axes[0].set(yticks=result.Year, yticklabels=result.Year)
axes[0].yaxis.tick_left()

axes[1].set_xticks([1e+09, 2e+09, 3e+09, 4e+09, 5e+09, 6e+09, 7e+09])
axes[1].spines['left'].set_visible(False)
axes[1].set_yticks([])

for label in (axes[0].get_xticklabels() + axes[0].get_yticklabels()):
    label.set(fontsize=13)
for label in (axes[1].get_xticklabels() + axes[1].get_yticklabels()):
    label.set(fontsize=13)

plt.subplots_adjust(wspace=0, top=0.85, bottom=0.1, left=0.18, right=0.95)

plt.savefig('People with and without access to electricity per year (bidirectional barchart with scale error).jpg', dpi=300, bbox_inches='tight')
    
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.stackplot(result.Year, result['Number of people with access to electricity'],
             result['Number of people without access to electricity'],
             colors=['#01b8aa', '#fd625e'], labels=['with access', 'without access'])
ax.legend(loc='upper left')
plt.margins(x=0)
ax.set_title('Number of people with and without acces to electricity', fontsize=15, pad=10)
ax.set_ylabel('number of people in billion', fontsize= 12)

ax.set_xticks([1999, 2003, 2007, 2011, 2015, 2019])

plt.savefig('People with and without access to electricity per year (area chart).jpg', dpi=300, bbox_inches='tight')

plt.show()
```

```python
with_access = []
without_access = []
years = []

for i in range(len(result)):
    with_access.append(float('{0:.2f}'.format(result['Number of people with access to electricity'].loc[i + 6474] / 1000000000)))
    without_access.append(float('{0:.2f}'.format(result['Number of people without access to electricity'].loc[i + 6474] / 1000000000)))
    years.append(result['Year'].loc[i + 6474])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

rects1 = ax.bar(years, with_access, color='#01b8aa', label='with access')
rects2 = ax.bar(years, without_access, color='#fd625e', label='without access')

ax.set_ylabel('Billion people', fontsize=12)
ax.set_title('Number of people with and without acces to electricity', fontsize=15, pad=10)
ax.set_xticks(years)
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.legend()

plt.margins(x=0.025)

fig.tight_layout()

plt.savefig('People with and without access to electricity per year (stacked barchart).jpg', dpi=300, bbox_inches='tight')

plt.show()
```

```python
x = np.arange(len(result))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width/2, with_access, width, label='with access', color='#01b8aa')
rects2 = ax.bar(x + width/2, without_access, width, label='without access', color='#fd625e')

ax.set_ylabel('Billion people', fontsize=12)
ax.set_title('Number of people with and without acces to electricity', fontsize=15, pad=10)
ax.set_xticks(x, years)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

plt.margins(x=0.025)
fig.tight_layout()

plt.savefig('People with and without access to electricity per year (barchart).jpg', dpi=300, bbox_inches='tight')

plt.show()
```
