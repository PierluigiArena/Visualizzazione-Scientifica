```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('number-without-electricity-by-region.csv')

result = df[df['Entity'].isin(['South Asia', 'Micronesia (country)',
                               'Latin America & Caribbean (IDA & IBRD)', 'European Union', 
                               'Europe & Central Asia (excluding high income)', 'East Asia & Pacific',
                               'Africa Western and Central', 'Africa Eastern and Southern'])]

result = result.drop(['Code'], axis=1)
```

```python
America = pd.DataFrame([], columns=['Entity', 'Year', 'Number of people'])
Asia = pd.DataFrame([], columns=['Entity', 'Year', 'Number of people'])
Oceania = pd.DataFrame([], columns=['Entity', 'Year', 'Number of people'])
Africa = pd.DataFrame([], columns=['Entity', 'Year', 'Number of people'])
Europa = pd.DataFrame([], columns=['Entity', 'Year', 'Number of people'])

for i in range(20):
    America.loc[-1] = ['America', int(2000 + i), float(result['Number of people without access to electricity'].loc[i + 2547])]
    America.index = America.index + 1
    America = America.sort_index()

for i in range(20):
    Asia.loc[-1] = ['Asia', int(2000 + i), float(
        result['Number of people without access to electricity'].loc[i + 4348] + 
        result['Number of people without access to electricity'].loc[i + 1486] +
        result['Number of people without access to electricity'].loc[i + 1219])]
    Asia.index = Asia.index + 1
    Asia = Asia.sort_index()
    
for i in range(20):
    Oceania.loc[-1] = ['Oceania', int(2000 + i), float(
        result['Number of people without access to electricity'].loc[i + 3026] + 
        result['Number of people without access to electricity'].loc[i + 1486])]
    Oceania.index = Oceania.index + 1
    Oceania = Oceania.sort_index()
    
for i in range(20):
    Africa.loc[-1] = ['Africa', int(2000 + i), float(
        result['Number of people without access to electricity'].loc[i + 35] + 
        result['Number of people without access to electricity'].loc[i + 15])]
    Africa.index = Africa.index + 1
    Africa = Africa.sort_index()
    
for i in range(20):
    Europa.loc[-1] = ['Europa', int(2000 + i), float(0)]
    Europa.index = Europa.index + 1
    Europa = Europa.sort_index()
```

```python
index = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019][::-1]

fig, ax = plt.subplots(figsize=(12, 8))
ax.stackplot(index, Europa['Number of people'], Africa['Number of people'],
             Oceania['Number of people'], Asia['Number of people'], America['Number of people'],
             colors=['royalblue', 'green', 'red', 'y', 'mediumvioletred'],
             labels=['Europe', 'Africa', 'Oceania', 'Asia', 'America'])
ax.legend(loc='upper left')
plt.margins(x=0)
ax.set_title('Country\'s number of people without acces to electricity per year', fontsize=15, pad=10)
ax.set_ylabel('number of people in billion', fontsize= 12)

plt.savefig('Country\'s people without access to electricity per year.jpg', dpi=300, bbox_inches='tight')

plt.show()
```
