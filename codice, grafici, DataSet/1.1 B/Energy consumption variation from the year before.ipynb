```python
import pandas as pd
import matplotlib.pyplot as plt

variazioni = pd.read_csv('cambiamenti nell\'uso dell\'energia.csv')
```

```python
result = variazioni[variazioni['Entity'] == 'World']
result = result.drop(['Entity', 'Code'], axis=1)
```

```python
plt.figure(figsize=(16,10))
plt.title("Variation of energy consumption per year", fontsize=20, pad=10)

plt.plot(result['Year'], result['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4)
plt.axhline(y=0, color='r', linestyle='-')
plt.grid(True)
plt.margins(x=0)
plt.ylabel('change of percentage compared to the previous year', fontsize=15)
plt.xticks(fontsize=14) 
plt.yticks(fontsize=14)
plt.savefig('Energy consumption variation from the previous year.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# questa variazione percentuale non indica altro che la variazione di consumo di energia tra un anno e un altro
# una variazione che passa da un valore positivo ad uno positivo minore indica che il consumo è aumentato ma meno rispetto
# all'aumento dell'anno precedente, una variazione che passa da un valore positivo ad uno positivo maggiore indica che il
# consumo è aumentato di più rispetto all'aumento di consumo dell'anno precedente.

# IMPORTANTE: variazioni che passano da valori positivi a valori negativi e viceversa indicano che il consumo totale 
#             rispetto all'anno precedente è diminuito o in caso contrario che è aumentato
```
