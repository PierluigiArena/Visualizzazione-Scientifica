```python
import pandas as pd
import matplotlib.pyplot as plt

variazioni = pd.read_csv('Energy consumption variation.csv')

result = variazioni[variazioni['Entity'] == 'World']
ita = variazioni[variazioni['Entity'] == 'Italy']
usa = variazioni[variazioni['Entity'] == 'United States']
afg = variazioni[variazioni['Entity'] == 'Afghanistan']
```

```python
result = result.drop(['Entity', 'Code'], axis=1)
ita = ita.drop(['Entity', 'Code'], axis=1)
usa = usa.drop(['Entity', 'Code'], axis=1)
afg = afg.drop(['Entity', 'Code'], axis=1)
```

```python
plt.figure(figsize=(16,10))
plt.title("Variation of energy consumption per year", fontsize=20, pad=10)

plt.plot(result['Year'], result['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4)
plt.plot(ita['Year'], ita['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4, color='g')
plt.plot(usa['Year'], usa['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4, color='m')

plt.axhline(y=0, color='r', linestyle='-')

plt.legend(['World', 'Italy', 'United States'])

plt.grid(True)
plt.margins(x=0)

plt.ylabel('percentage variation compared to the previous year', fontsize=15)

plt.xticks(fontsize=14) 
plt.yticks(fontsize=14)

plt.savefig('Energy consumption variation from the previous year (Italy, United States).jpg', dpi=300, bbox_inches='tight')

plt.show()
```

```python
new_result = result[result['Year'] >= 1981]
new_ita = ita[ita['Year'] >= 1981]

plt.figure(figsize=(16,10))
plt.title("Variation of energy consumption per year", fontsize=20, pad=10)

plt.plot(new_result['Year'], new_result['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4)
plt.plot(new_ita['Year'], new_ita['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4, color='g')
plt.plot(afg['Year'], afg['Annual change primary energy consumption (%)'], marker="o", linewidth=2, markersize=4, color='m')

plt.axhline(y=0, color='r', linestyle='-')

plt.legend(['World', 'Italy', 'Afghanistan'])

plt.grid(True)
plt.margins(x=0)

plt.ylabel('change of percentage compared to the previous year', fontsize=15)

plt.xticks(fontsize=14) 
plt.yticks(fontsize=14)

plt.savefig('Energy consumption variation from the previous year (Italy, Afghanistan).jpg', dpi=300, bbox_inches='tight')

plt.show()
```
